```python
import pandas as pd
ted = pd.read_csv('ted_main.csv')
transcripts = pd.read_csv('transcripts.csv')
ted.head()
```

```python
df = ted.merge(transcripts, on='url')
df.head()
```

```python
df = df[['title', 'description', 'main_speaker', 'tags', 'transcript', 'views']]
```

```python
df['description'] = df['description'].fillna('')
df['transcript'] = df['transcript'].fillna('')
df['tags'] = df['tags'].fillna('')
```

```python
import ast

def clean_tags(text):
    try:
        return " ".join(ast.literal_eval(text))
    except:
        return ""

df['tags'] = df['tags'].apply(clean_tags)
```

```python
df['content'] = (
    df['description'] + " " +
    df['transcript'] + " " +
    df['tags'] + " " +
    df['main_speaker']
)
```

```python
print(ted.shape)
print(transcripts.shape)
```

```python
print(df.shape)
df.head()
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    words = text.split()
    words = [w for w in words if w not in stop_words]
    words = [lemmatizer.lemmatize(w) for w in words]
    
    return " ".join(words)
```

```python
df['content'] = df['content'].apply(preprocess)
```

```python
df['content'][0][:300]
```

```python
df = df.reset_index(drop=True)
indices = pd.Series(df.index, index=df['title']).drop_duplicates()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=40000, ngram_range=(1,2))
tfidf_matrix = tfidf.fit_transform(df['content'])
print(tfidf_matrix.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_talk(title, top_n=5):
    
    if title not in indices:
        return ["Talk not found"]
    
    idx = indices[title]
    
    similarity = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
    
    weighted_score = (
        0.8 * similarity +
        0.2 * (df['views'] / df['views'].max())
    )
    
    similar_idx = weighted_score.argsort()[::-1][1:top_n+1]
    
    return df['title'].iloc[similar_idx].values
```

```python
recommend_talk("Do schools kill creativity?", 5)
```

```python
def recommend_talk(title, top_n=5):
    
    if title not in indices:
        return ["Talk not found"]
    
    idx = indices[title]
    
    similarity = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
    
    weighted_score = (
        0.8 * similarity +
        0.2 * (df['views'] / df['views'].max())
    )
    
    similar_idx = weighted_score.argsort()[::-1][1:top_n*2]  # take more initially
    
    results = df[['title','main_speaker','views']].iloc[similar_idx]
    
    results = results.drop_duplicates(subset='title')  # REMOVE DUPLICATES
    
    return results.head(top_n)
```

```python
recommend_talk("The power of vulnerability", 5)
```

```python
import pickle

pickle.dump(tfidf, open('tfidf.pkl','wb'))
pickle.dump(tfidf_matrix, open('matrix.pkl','wb'))
pickle.dump(indices, open('indices.pkl','wb'))
df.to_pickle('data.pkl')
```
